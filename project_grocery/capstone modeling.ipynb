```python
import dask.dataframe as dd

# Define the expected data types for columns
dtype = {
    'Unnamed: 0': 'int64',
    'item_nbr': 'int64',
    'store_nbr': 'int64',
    'unit_sales': 'float64'
}

# Read the CSV file into a Dask DataFrame with specified data types
df = dd.read_csv('all_info.csv', dtype=dtype)

# Assuming `dask_df` is your Dask DataFrame
sample_df = df.compute()
```

```python
import pandas as pd

# Calculate the fraction to sample 12 million rows 
frac = 1000000 / len(df)

# Randomly sample 1 million rows
sampled_df = sample_df.sample(frac=frac, random_state=42)

# Reset the index if needed
sampled_df = sampled_df.reset_index(drop=True)
sampled_df = sampled_df.loc[sampled_df['unit_sales'] >= 0]
```

```python
import pandas as pd
# Define the categorical columns
categorical_columns = [
    'onpromotion', 'type_x', 'type_y', 'cluster', 'city', 
    'family', 'perishable', 'class', 'state', 'year', 
    'transferred', 'locale', 'month', 'day', 'locale_name'
]
df_encoded = pd.get_dummies(sampled_df, columns=categorical_columns, drop_first=True)
#df_merged = pd.concat([sampled_df, df_encoded], axis=1)
```

```python
df_merged=df_encoded
```

```python
from sklearn.model_selection import train_test_split
X=df_merged.drop(columns=['unit_sales','date'])
y=df_merged['unit_sales']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error
import numpy as np

# Subsample the data for quicker hyperparameter tuning
X_train_sampled = X_train.sample(frac=0.5, random_state=42)
y_train_sampled = y_train.loc[X_train_sampled.index]

# Define a smaller parameter distribution
param_dist = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['auto', 'sqrt'],
    'bootstrap': [True, False]
}

# Initialize Random Forest
rf = RandomForestRegressor(random_state=42, warm_start=True)

# Randomized Search CV
rf_random = RandomizedSearchCV(estimator=rf, 
                               param_distributions=param_dist, 
                               n_iter=20,  # Reduced iterations
                               cv=3, 
                               verbose=2, 
                               random_state=42, 
                               n_jobs=-1, 
                               scoring='neg_root_mean_squared_error')

# Fit the model on the sampled data
rf_random.fit(X_train_sampled, y_train_sampled)

# Print the best hyperparameters found by RandomizedSearchCV
print(f"Best Hyperparameters: {rf_random.best_params_}")

# Use the best model on the full training set
best_rf = rf_random.best_estimator_

# Predict on both training and test sets
y_pred_train = best_rf.predict(X_train)
y_pred_test = best_rf.predict(X_test)

# Calculate RMSE for both training and test sets
train_rmse = mean_squared_error(y_train, y_pred_train, squared=False)
test_rmse = mean_squared_error(y_test, y_pred_test, squared=False)

# Print RMSE results
print(f"Training RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import RandomizedSearchCV, train_test_split
import numpy as np

# Sample a fraction of the data for quicker hyperparameter tuning
X_train_sampled, _, y_train_sampled, _ = train_test_split(X_train, y_train, test_size=0.5, random_state=42)

# Flatten y if needed
y_train_sampled = y_train_sampled
y_test = y_test

# Initialize HistGradientBoosting Regressor
hgbr = HistGradientBoostingRegressor(random_state=42)

# Define a smaller parameter distribution for quicker tuning
param_dist = {
    'learning_rate': [0.01, 0.1],
    'max_iter': [100, 200],
    'max_depth': [5, 7],
    'min_samples_leaf': [20, 30],
    'max_bins': [255],
}

# Randomized Search CV
hgbr_random = RandomizedSearchCV(estimator=hgbr, 
                                 param_distributions=param_dist, 
                                 n_iter=5,  # Reduced number of iterations
                                 cv=3, 
                                 verbose=2, 
                                 random_state=42, 
                                 n_jobs=-1, 
                                 scoring='neg_root_mean_squared_error')

# Fit the model on the sampled data
hgbr_random.fit(X_train_sampled, y_train_sampled)

# Get the best estimator
best_hgbr = hgbr_random.best_estimator_

# Make predictions
y_pred_train = best_hgbr.predict(X_train)
y_pred_test = best_hgbr.predict(X_test)

# Calculate RMSE
train_rmse = mean_squared_error(y_train, y_pred_train, squared=False)
test_rmse = mean_squared_error(y_test, y_pred_test, squared=False)

print(f"Best Hyperparameters: {hgbr_random.best_params_}")
print(f"Training RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")

```

# We will be using the  HistGradientBoost Regressor Model as it has a lower RMSE and the Test RMSE and Train RMSE are closer together than the  RandomForest Regressor Model
